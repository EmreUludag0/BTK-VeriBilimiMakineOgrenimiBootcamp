```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression #dogrusal
from sklearn.preprocessing import PolynomialFeatures # polinomsal
from sklearn.metrics import r2_score # r kare
```

```python
df = pd.read_csv("dogrusal_regresyon_veriseti_20220209.csv", sep=";")
df.head()
```

```python
# data icerisinde bulunan maas ve deneyim sutunlari numpy array'e cevirir
x = df.deneyim.values.reshape(-1,1)
y = df.maas.values.reshape(-1,1)
```

```python
# Dogrusal Regresyon Modeli
linear_reg = LinearRegression()
# Dogrusal Regresyon Egitimi
linear_reg.fit(x,y)
```

```python
# Polinom regresyon modeli Egitimi
polinom_regresyon = PolynomialFeatures(degree = 2) # 2. dereceden polinom

x_polinom = polinom_regresyon.fit_transform(x)

poly_reg = LinearRegression()
poly_reg.fit(x_polinom, y)
```

```python
#test 
y_tahmin_linear = linear_reg.predict(x)
y_tahmin_poly = poly_reg.predict(x_polinom)

plt.scatter(df.deneyim, df.maas)
plt.plot(x,y_tahmin_linear,color = "red", label= "dogrusal")
plt.scatter(x, y_tahmin_poly, color = "green", label = "polinomsal")
plt.xlabel("deneyim(yil)")
plt.ylabel("maas(TL)")
plt.title("deneyim maas ilişkisi")
plt.grid(True)
plt.show()
```

```python
print("dogrusal regresyon R kare:{}", r2_score(y,y_tahmin_linear))
print("polinomsal regresyon R kare:{}",r2_score(y,y_tahmin_poly))
```
