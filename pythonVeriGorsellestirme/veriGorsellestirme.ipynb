```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmis_20220204.csv")
```

Saçılım Grafigi

```python
def sacilimGrafikCiz():
    sns.scatterplot(x='boy', y='kilo', data=veri)
    plt.title("Boy ve Agirlik Iliskisi - Izgara Tema")
    plt.show()

sacilimGrafikCiz()
```

```python
sns.set_style("white")
sacilimGrafikCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

Sacilim Grafigi - Kategorik

```python
#örnek 1
sns.set_style("darkgrid")
sns.scatterplot(x='boy', y='kilo', hue='madalya', data=veri)
plt.title('Madalyaya Göre Boy ve Agirlik Dagilimi')
plt.show()
```

```python
#örnek2
sns.set_style("darkgrid")
sns.scatterplot(x='boy',y='kilo',hue='Gender', data=veri)
plt.title("Cinsiyete Göre Dagilim")
plt.show()
```

```python
#örnek3 (4 boyutlu görselleştirme)
sns.set_style("darkgrid")
sns.scatterplot(x='boy',y='kilo',hue='Gender', style='madalya', data=veri)
plt.title("Cinsiyete Göre Dagilim")
plt.show()
```

```python
#örnek4 (5 boyutlu görselleştirme)
sns.set_style("darkgrid")
sns.scatterplot(x='boy',y='kilo',hue='Gender', style='madalya',size='yas',sizes=(15,200), data=veri)
plt.title("Cinsiyete Göre Dagilim")
plt.show()
```

Saçilim Grafigi - Dogrusal Regresyon

```python
sns.regplot(x='boy',y='kilo', data=veri, marker='+', scatter_kws={'alpha':0.2})
plt.title('Boy ve Agirlik Dagilimi')
plt.show()
```

Renk Paletleri

```python
#örnek1
sns.scatterplot(x='boy',y='kilo', hue='madalya',data=veri, palette='rocket')
plt.title('Madalya Göre Boy ve Agirlik Dagilimi')
plt.show()
```

```python
#örnek2
sns.scatterplot(x='boy',y='kilo', hue='madalya',data=veri, palette='Set1')
plt.title('Madalya Göre Boy ve Agirlik Dagilimi')
plt.show()
```

Çizgi Grafigi

```python
sns.lineplot(x='boy', y='kilo', data=veri)
plt.title('Boy ve Kilo')
plt.show()
```

Cizgi Grafigi - Kategorik

```python
sns.lineplot(x='boy',y='kilo',hue='Gender',data=veri)
plt.title("Cinsiyete Gore Boy ve Agirlik Iliskileri")
plt.show()
```

Histogram

```python
sns.displot(veri, x='kilo')
plt.ylabel('frekans')
plt.title('Agirlik Histogrami')
plt.show()
```

Histogram - Kategorik

```python
sns.displot(veri, x='kilo',hue="Gender")
plt.ylabel("frekans")
plt.title("Cinsiyete Gore Dagilim")
plt.show()
```

yan yana iki sutun olustur

```python
sns.displot(veri, x="kilo",col='Gender', multiple="dodge")
plt.show()
```

iki boyutlu histogram (izohips Haritasi)

```python
sns.displot(veri, x="kilo", y="boy", kind="kde")
plt.xlabel("agirlik")
plt.ylabel("Boy")
plt.title("Agirlik - Boy Histogrami")
plt.show()
```

```python
sns.displot(veri, x="kilo", y="boy",hue="Gender", kind="kde")
plt.xlabel("agirlik")
plt.ylabel("Boy")
plt.title("Cinsiyete Gore Agirlik - Boy Histogrami")
plt.show()
```

Cubuk Grafigi

```python
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("madalyalara Boy Grafigi")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", data=veri, hue="Gender")
plt.title("madalyalara Boy Grafigi")
plt.show()
```

```python
sns.catplot(x="madalya",y="yas", hue="Gender", col="sezon", data=veri, kind="bar", height=4, aspect=0.7)
plt.show() 
```

Kutu Grafigi

```python
sns.boxplot(x="sezon",y="boy",data=veri)
plt.show()
```

Keman Grafigi

```python
sns.violinplot(x="sezon", y="boy",data=veri)
plt.show()
```

Cift/Es Grafigi

```python
sns.pairplot(data=veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde= True)
```
