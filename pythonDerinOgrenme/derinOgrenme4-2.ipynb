## ızgara aramasi capraz dogrulama
* makine ogrenmesinde hiperparametre, degeri ogrenme surecini kontrol etmek icin kullanılan bir parametredir.
* ornek, k-en yakın komsu algoritmasında kullanılan k degeri bir hiperparametredir. bu hiperparametrenin secılması gereklidir.
* ızgara aramsı çapraz dogrulama modelin basarımını dogrudan etlişeyen hiperparametrelerin seciminde kullanılır.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
#veri yükleme
iris = load_iris()

x = iris.data # özellikler
y = iris.target # sinif
```

```python
#normalizasyon 
x = (x - np.min(x))/(np.max(x)-np.min(x))
```

```python
#egitim test bölmesi
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) # k = n_neighbors
```

```python
from sklearn.model_selection import GridSearchCV

grid = {"n_neighbors":np.arange(1,50)} 
knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn, grid, cv = 10) # gridsearchcv
knn_cv.fit(x,y)
```

```python
print("en iyi k degeri:", knn_cv.best_params_)
print("en iyi k degerine gore en iyi dogruluk degeri: ", knn_cv.best_score_)
```

lojistik regresyon modeli için en iyi parametreleri bulalım

```python
from sklearn.linear_model import LogisticRegression

grid = {"C":np.logspace(-3,3,7),"penalty":["l1","l2"]} 
logreg = LogisticRegression()
logreg_cv = GridSearchCV(logreg,grid,cv = 10)
logreg_cv.fit(x,y)

print("en iyi hiper parametreler:", logreg_cv.best_params_)
print("en iyi hiper parametlere göre en iiyi dogrluk degeri:", logreg_cv.best_score_)
```
